```python
#Convert createdDate and Closed Date to dateTime
from datetime import datetime
dataset_nyc['Created Date'] = pd.to_datetime(dataset_nyc['Created Date']).dt.tz_localize(None)
dataset_nyc['Closed Date'] = pd.to_datetime(dataset_nyc['Closed Date']).dt.tz_localize(None)

dataset_nyc.dtypes[['Created Date','Closed Date']]
```

```python
#Create a column Request_Closing_Time with value = 'Closed Date'  - 'Created Date'
from datetime import datetime
dataset_nyc['Request_Closing_Time'] = dataset_nyc['Closed Date'] - dataset_nyc['Created Date']
print(dataset_nyc['Request_Closing_Time'].head(10))
# Convert the Request_Closing_Time to hours
dataset_nyc['Request_Closing_Time'] = ((dataset_nyc['Request_Closing_Time'].dt.seconds/60)/60)
dataset_nyc['Request_Closing_Time'].fillna(0,inplace=True)
dataset_nyc['Request_Closing_Time'] = dataset_nyc['Request_Closing_Time'].astype(int)

#print(dataset_nyc['Request_Closing_Time'].head(10))
dataset_nyc.head(5)
```

```python
# Are the type of complaint or service requested and location related?
##preparing dataframe and group for 'Complaint Type', 'city'
df_gptest2 = dataset_nyc[['Complaint Type', 'City']].dropna()

grouped_city=df_gptest2[['Complaint Type', 'City']].groupby(['Complaint Type'])
print(len(grouped_city)) # Displays one record per group

print(grouped_city.head()) # Displays one record per group
```

```python
#chi-Square testing is used when both the varibales are categorical.
from scipy.stats import chi2_contingency
import scipy.stats

scipy.stats.chisquare(df_gptest2['Complaint Type'].value_counts())

scipy.stats.chisquare(df_gptest2['City'].value_counts())

cont = pd.crosstab(df_gptest2['Complaint Type'],
                   df_gptest2['City'])

scipy.stats.chi2_contingency(cont)

## Null Hypothesis- there's no correlation between complaint type and City
## p-val is 0.0 suggesting high correlation between complaint type and the City, 
#we reject the null hpyothesis.   
```
